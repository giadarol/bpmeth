```python
import numpy as np
import bpmeth
import xtrack as xt
import xobjects as xo
```

```python
env = xt.Environment()
env.particle_ref = xt.Particles(p0c=7e12)

# Fringe in straight frame
h = 0
length = 5

drift = bpmeth.DriftVectorPotential()
H_drift = bpmeth.Hamiltonian(length, h, drift)

line1 = env.new_line(components=[
    env.place('fringe', H_drift, at=0),
    ])

line2 = env.new_line(components=[
    env.new('drift', xt.Drift, length=length)
])

tw1 = line1.twiss(betx=1, bety=1, include_collective=True)
tw2 = line2.twiss(betx=1, bety=1, include_collective=True)
```

```python
tw1.plot()
tw2.plot()
```

```python
env = xt.Environment()
env.particle_ref = xt.Particles(p0c=7e12)

# Fringe in straight frame
h = 0
b1 = 1
aa = 1
length = 5*aa
b1fringe = f"{b1}*(tanh(s/{aa})+1)/2"
# Fringe field integral K = 1/(2g)
fringe = bpmeth.FringeVectorPotential(b1fringe)

H_fringe = bpmeth.Hamiltonian(length, h, fringe)


line = env.new_line(components=[
    env.place('fringe', H_fringe, at=0),
    ])

tw = line.twiss(betx=1, bety=1, include_collective=True)
```

```python
tw.plot()
```
